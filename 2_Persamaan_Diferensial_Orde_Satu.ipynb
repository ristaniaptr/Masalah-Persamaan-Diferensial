# **Nomor 1**

Misalkan pertumbuhan populasi ikan di suatu perairan digambarkan oleh persamaan

$$
\dfrac{dy}{dt} = r\left(1-\dfrac{y}{K}\right)y-Ey
$$

dengan $r \in \mathbb{R}$ menyatakan laju pertumbuhan ikan, $K \in \mathbb{R}$ menyatakan konstanta daya dukung lingkungan, dan $E \in \mathbb{R}$ menyatakan laju penangkapan ikan.



## **Nomor 1A**

Tentukan titik kesetimbangan dari persamaan diferensial tersebut jika


*   $E < r$
*   $E = r$
*   $E > r$



**Jawab:**

Perhatikan bahwa jika

$$
\dfrac{dy}{dt} = r\left(1-\dfrac{y}{K}\right)y-Ey = y\left[r\left(1-\dfrac{y}{K}\right)-E\right]=0,
$$

maka diperoleh dua titik kesetimbangan, yaitu
$$
y = 0, y = K\left(1-\dfrac{E}{r}\right).
$$

Dengan demikian,
*   Jika $E < r$, terdapat 2 titik kesetimbangan yaitu $y = 0$ dan $y = K\left(1-\dfrac{E}{r}\right)>0$
*   Jika $E = r$, hanya terdapat 1 titik kesetimbangan yaitu $y = 0$
*   Jika $E > r$, didapat $y = 0$ dan $y = K\left(1-\dfrac{E}{r}\right)<0$. Namun karena $y$ menyatakan banyak populasi ikan, maka nilai $y$ tidak mungkin negatif sehingga titik kesetimbangannya hanyalah $y = 0$.

## **Nomor 1B**

Untuk $r = 2$, $K = 10$, dan $E = 1.5$, gambarkan medan gradien dari persamaan diferensial tersebut bersamaan dengan beberapa kurva solusinya jika diketahui nilai awal $y(0) = -1, 1, 4$.

```python
from sympy import *
from sympy.abc import t
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
y = Function('y')
# Ketik jawaban Anda di bawah garis.
# -----------------------------------------------------------------------

# Parameter
r = 2
K = 10
E = 1.5

# Menggambar Beberapa Kurva Solusi
sol = dsolve(Derivative(y(t), t) - r*(1-y(t)/K)*y(t) + E*y(t), y(t)).rhs
# untuk c = -1
constant1 = solve([sol.subs(t, 0) - (-1)])
final_sol1 = sol.subs(constant1)
final_sol_func1 = lambdify(t, final_sol1)
# untuk c = 1
constant2 = solve([sol.subs(t, 0) - 1])
final_sol2 = sol.subs(constant2)
final_sol_func2 = lambdify(t, final_sol2)
# untuk c = 4
constant3 = solve([sol.subs(t, 0) - 4])
final_sol3 = sol.subs(constant3)
final_sol_func3 = lambdify(t, final_sol3)

t = np.linspace(0, 2, 10)
y = np.linspace(-5, 5, 10)

# Definisikan Persamaan Diferensial
def dydt(t,y) :
     return r*(1-y/K)*y - E*y

# Gridisasi
ts, ys = np.meshgrid(t,y)
plt.scatter(ts,ys)

# Menggambar Medan Gradien
for j in t:
  for k in y:
    slope = dydt(j,k)
    domain = np.linspace(j-0.05, j+0.05, 2)
    def fun(t1,y1):
      z = slope*(domain - t1) + y1
      return z
    plt.plot(domain, fun(j,k), 'k', solid_capstyle = 'projecting', solid_joinstyle = 'bevel')

t_s = np.linspace(0,2,100)
plt.plot(t_s, final_sol_func1(t_s))
plt.plot(t_s, final_sol_func2(t_s))
plt.plot(t_s, final_sol_func3(t_s))
plt.xlim([0,2])
plt.ylim([-5,5])
plt.xlabel('$t$')
plt.ylabel('$y(t)$')
plt.grid('on')
plt.show()
```

## **Nomor 2A**

Dengan menggunakan teorema eksistensi dan ketunggalan, apakah MNA tersebut memiliki solusi untuk $a=-1, 0, 1$? Jika ya, apakah solusinya tunggal?

**Jawab:**

*   Untuk $a = -1, y(1) = -1$

$f(x,y) = \frac{2y}{(x-1)^2}$ tidak kontinu di suatu persegi panjang yang memuat (1,-1), maka MNA ini tidak dijamin memiliki solusi.


*   Untuk $a = 0, y(0) = 0$

$f(x,y) = \frac{2y}{(x-1)^2}$ kontinu di suatu persegi panjang yang memuat (0,0), maka MNA ini memiliki solusi.

$f_y(x,y) = \frac{2}{(x-1)^2}$  kontinu di suatu persegi panjang yang memuat (0,0), maka MNA ini memiliki solusi tunggal.


*   Untuk $a = 1, y(-1) = 1$

$f(x,y) = \frac{2y}{(x-1)^2}$ kontinu di suatu persegi panjang yang memuat (-1,1), maka MNA ini memiliki solusi.

$f_y(x,y) = \frac{2}{(x-1)^2}$  kontinu di suatu persegi panjang yang memuat (-1,1), maka MNA ini memiliki solusi tunggal.

## **Nomor 2B**

Jika MNA tersebut eksistensi solusinya terjamin, tentukan semua solusi umum dan khususnya dengan menggunakan paket Sympy.

Ketika $a=0$, apa yang terjadi?

```python
from sympy import *
from sympy.abc import x
y = Function('y')
# Ketik jawaban Anda di bawah garis.
# -----------------------------------------------------------------------

# Mencari Solusi Umum
sol = dsolve(Derivative(y(x),x) - 2*y(x)/(x-1)**2, y(x))
sol
```

```python
# Mencari Solusi Khusus untuk a = 0

constant0 = solve([sol.rhs.subs(x,0) - 0])
print(constant0)
final_ans0 = sol.rhs.subs(constant0)
print('Solusi khusus untuk a = 0: y(x) = ', final_ans0)
```

Ketika $a=0$, solusi MNA tersebut adalah $y=0$. Sayangnya, `dsolve` memiliki keterbatasan ketika bekerja dengan ODE yang solusinya trivial.

```python
# Mencari Solusi Khusus untuk a = 1
constant1= solve([sol.rhs.subs(x,-1) - 1])
print(constant1)
final_ans1 = sol.rhs.subs(constant1)
print('Solusi khusus untuk a = 1: y(x) = ', final_ans1)
```
